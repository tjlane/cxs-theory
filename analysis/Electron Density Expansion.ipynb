```python
from odin import xray
import mdtraj
from odin.refdata import sph_quad_900
from odin.math2 import arctan3

from scipy import misc, special
```

```python
#s = mdtraj.load('pentagon/pentagon.pdb')
s = mdtraj.load('octagon/octagon.pdb')
print s.xyz
```

```python
def gaussian_density(x, x0, sigma):
    d = np.exp( - np.sum(np.square( x - x0 ) / (2.0 * sigma**2), axis=1) )
    N = np.power( 2.0 * sigma**2 * np.pi, -3.0/2.0 )
    return N * d

def electron_density(positions, structure=s):
    """
    Compute an electron density scalar field evaluated at `positions`
    """
    sigma = 0.05
    
    density = np.zeros(positions.shape[0])
    assert positions.shape[1] == 3
    
    for atom in s.xyz[0]:
        #print gaussian_density(positions, atom[None,:], sigma)
        density += gaussian_density(positions, atom[None,:], sigma)
    
    return density
```

```python
l_vals = 30
r_max = 1.0
radii_samples = arange(0, r_max, 0.01)


# project rho onto the spherical harmonics using spherical quadrature
phi = arctan3(sph_quad_900[:,1], sph_quad_900[:,0])
g_lm = np.zeros((l_vals, 2*l_vals+1, len(radii_samples)), dtype=np.complex128)

for ir,r in enumerate(radii_samples):
    
    rho_r = electron_density(r * sph_quad_900[:,:3], structure=s)
    
    for il,l in enumerate(range(l_vals)):                
        for m in range(-l, l+1):
            
            k = m + l# reindex
            
            N = np.sqrt( 2. * l * misc.factorial(l-m) / \
                       ( 4. * np.pi * misc.factorial(l+m) ) )
            Plm = special.lpmv(m, l, sph_quad_900[:,2])
            Ylm = N * np.exp( 1j * m * phi ) * Plm

            g_lm[il, k, ir] = np.sum( rho_r * Ylm * sph_quad_900[:,3] )
```

```python
figure()

imshow( np.sum(np.square(np.abs(g_lm)), axis=1), aspect='auto', interpolation='nearest',
        extent=[0,r_max,l_vals,0] )
xlabel('radius (nm)')
ylabel(r'$\ell^{2}$ Coefficient')

show()
```

```python
g_lm_max = np.abs(g_lm[:,:,radii_samples == 0.4])

img = np.squeeze(g_lm_max)

figure(figsize=(12,5))

subplot(121)
imshow(img, interpolation='nearest', aspect='auto')
ylabel(r'$\ell$ angular quantum number')
xlabel('m magnetic quantum number')

subplot(122)
plot( np.sum(np.square(np.abs(g_lm)), axis=1)[:,radii_samples == 0.4], np.arange(img.shape[0]) )
ylim([30,0])
xlabel('sum-squared intensity')

show()
```
