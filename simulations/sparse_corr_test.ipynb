```python
def compute_h(l, L, m, M, lbd):
    N = np.sqrt( (2.0*l + 1.0)*(2.0*L + 1.0)*(2.0*lbd + 1.0) / (4.0*np.pi) ) *\
        Wigner3j(l, L, lbd, 0.0, 0.0, 0.0)
    return (-1.0)**M * N * Wigner3j(l, L, lbd, m, -M, M-m)


def forward_index(ell, m):
    """
    (ell, m) --> k, k = 0,1,2,...
    """
    return int(ell * (ell+1) + m)


def backward_index(k):
    """
    k --> (ell, m)
    
    Returns ell, m
    """
    ell = int(np.floor(np.sqrt(k)))
    m = int(k - ell*(ell+1))
    return ell, m


# indexing functions for (ell, m) --> k, k = 0,1,2,...
def forward_index_compact(ell, m):
    return int( (ell * (ell+1))/2 + m)


# and k --> (ell, m)
def backward_index_compact(k):
    """
    Returns ell, m
    """
    ell = int(np.floor( (np.sqrt(8*k+1) - 1)/2 ))
    m = int( k - (ell*(ell+1))/2 )
    return ell, m


def compute_H(lbd):
    """
    Return the matrix H_(kK)
    """
    n = (lbd ** 2) + 2 * lbd
    H = zeros((n, n))
    
    for k in range(n):
        for K in range(k, n):
            
            l, m = backward_index(k)
            L, M = backward_index(k)
            
            h = compute_h(l, L, m, M, lbd)
            
            H[k,K] = h
            H[K,k] = h
    
    return H


def compute_trunc_H(lbd):
    """
    Return the matrix H_(kK)^trunc
    """
    n = (lbd ** 2) + 2 * lbd
    H = zeros((n, n))
    
    for k in range(n):
        for K in range(k, n):
            
            l, m = backward_index(k)
            L, M = backward_index(k)
            
            h = compute_h(l, L, m, M, lbd)
            
            H[k,K] = h
            H[K,k] = h
    
    return H


def compute_lsqd_coef(Gk):
    """
    Take the output from the optimizer (which has the real & complex components separated)
    and compute the ell-squared coefficient,
    
        ell-squared = sum{ G_{ell m} ^ 2 }
                       m
                       
    """
    
    Gk_mag = np.square( np.abs(Gk) ) # k-indexing now correct
    k_max = len(Gk_mag)
    
    l_max = backward_index(k_max)[0]
    lsqd = np.zeros(l_max+1)
    
    for k in range(k_max):
        l,m = backward_index(k)
        lsqd[l] += Gk_mag[k]
        
    return lsqd


def plot_eigenspectrum(matrix):
    
    lbds = np.linalg.eig(matrix)[0]
    
    figure(figsize=(1,8))
    ax = subplot(111)
    hlines(lbds, 0, 1, lw=2)
    xlim([0,1])
    ax.set_xticklabels([])
    show()
    return


def C_lbd_to_M(c_vector):
    d = np.array([2.0 * l + 1.0 for l in range(len(c_vector))])
    M = np.sqrt( np.abs(c_vector) / d)
    return M / M[1]
```

```python
# compose H tensor
print 'lbd_max:', lbd_max

nm = (lbd_max ** 2) + 2 * lbd_max
print nm
HT = np.zeros((lbd_max, nm, nm))

for lbd in range(lbd_max):
    n = (lbd ** 2) + 2 * lbd
    HT[lbd,:n,:n] = compute_H(lbd)
```

```python
n_param = (((lbd_max+1) ** 2) + 3 * (lbd_max+1))/2


def Gparam_to_Gk(G_param):
    
    G_param = G_param.reshape(n_param, 2) # split real and imag
    G_param = G_param.astype(np.complex128)
    
    Gk_v = np.zeros(nm, dtype=np.complex128)
    Gk_s = np.zeros(nm, dtype=np.complex128)
    
    for k in range(nm):
        
        l,m = backward_index(k)
                
        if m >= 0:
            k_comp = forward_index_compact(l,m)
            Gk_v[k] = np.copy(G_param[k_comp,0] + 1j*G_param[k_comp,1]) # coefficient
            Gk_s[k] = np.copy(G_param[k_comp,0] - 1j*G_param[k_comp,1]) # complex conjugate
            
        else:
            k_comp = forward_index_compact(l, np.abs(m))
            # note switch in complex conj
            sign = -1 ** int(np.abs(m))
            Gk_v[k] = sign * np.copy(G_param[k_comp,0] - 1j*G_param[k_comp,1]) # complex conjugate
            Gk_s[k] = sign * np.copy(G_param[k_comp,0] + 1j*G_param[k_comp,1]) # non conj coeff
            
    return Gk_v, Gk_s


def sigma_gk(Gk_v):
    sigma = np.zeros_like(Gk_v)
    sigma[ Gk_v > 0.0 ] =  1.0
    sigma[ Gk_v < 0.0 ] = -1.0
    sigma[ Gk_v == 0.0 ] =  0.0
    return sigma
    
    
def objective(G_param, alpha=0.1):
    
    Gk_v, Gk_s = Gparam_to_Gk(G_param)
    
    obj = 0.0
    for lbd in range(lbd_max): 
        obj += np.power( np.abs( np.dot(Gk_s, np.dot(HT[lbd,:,:], Gk_v)) - M_lbd[lbd] ), 2)

    #print obj, np.sum(np.abs(Gk_v) * alpha)
    obj += np.sum(np.abs(Gk_v) * alpha)
    
    return obj


def objective_grad(G_param, alpha=0.1):
    
    Gk_v, Gk_s = Gparam_to_Gk(G_param)
    
    grad = np.zeros((len(Gk_v), 2))
    
    for lbd in range(lbd_max):
        g = 4.0 * np.power( np.abs( np.dot(Gk_s, np.dot(HT[lbd,:,:], Gk_v)) - M_lbd[lbd] ), 2) * \
            np.dot(HT[lbd,:,:], Gk_v) + alpha * sigma_gk(Gk_v)
        grad[:,0] += np.real(g)
        grad[:,1] += np.imag(g)
        
    # translate to compact indices
    grad_compact = np.zeros((n_param, 2))
    for k in range(grad.shape[0]):
        l,m = backward_index(k)
        if m >=0:
            #print k, l, m, forward_index_compact(l,m)
            grad_compact[forward_index_compact(l,m),:] = grad[k,:]
        
    return grad_compact.flatten()
```

```python
from scipy import optimize

n_param = (((lbd_max+1) ** 2) + 3 * (lbd_max+1))/2
Gk0 = np.random.randn(n_param*2) / 10. # x2 for complex
#Gk_opt = optimize.fmin(objective, Gk0, maxfun=int(1e6), maxiter=int(1e6), disp=1)
Gk_opt = optimize.fmin_bfgs(objective, Gk0, fprime=objective_grad, maxiter=int(1e6), disp=1)

Gk_v, Gk_s = Gparam_to_Gk(Gk_opt)
```
